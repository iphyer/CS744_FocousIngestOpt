```python
import re
r = re.compile("\d+")
def cussort(a, b):
    m1 = r.findall(a)
    m2 = r.findall(b)
    if m1[0] > m2[0]:
        return 1
    elif m1[0] < m2[0]:
        return -1
    else:
        if m1[1] >= m2[1]:
            return 1
        else:
            return -1
```
